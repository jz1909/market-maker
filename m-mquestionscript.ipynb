```python
Question = Dict[str, Any]
```

```python
session = requests.Session()
retries = Retry(
    total=5,
    backoff_factor=0.5,              # 0.5s, 1s, 2s, 4s...
    status_forcelist=[429, 500, 502, 503, 504],
    allowed_methods=["GET"],
    raise_on_status=False,
)
session.mount("https://", HTTPAdapter(max_retries=retries))

def fetch_json(url: str):
    resp = session.get(url, timeout=(5, 20))  # connect timeout 5s, read timeout 20s
    resp.raise_for_status()
    return resp.json()
```

```python
def world_bank_pop(country3: str) -> Dict[str, Any]:
    """
    Fetch the most recent non-null value for a World Bank indicator.

    Example endpoint:
    https://api.worldbank.org/v2/country/USA/indicator/SP.POP.TOTL?format=json
    """
    source_url = (
          f"https://api.worldbank.org/v2/country/{country3}/indicator/SP.POP.TOTL?format=json"
    )

    data = fetch_json(source_url)
    if data is None or len(data) < 2 or data[1] is None:
        return None

    observations: List[Dict[str, Any]] = data[1]

    return {
        "observations": observations,
        "sourceUrl": source_url,
    }
    
def run_world_bank_pop():
        
        out: List[Question] = []
        # countries = ["USA"]
    
        for country in countries:
       
            result = world_bank_pop(country)
            print(country)
            if result is None:
                continue
                
            c_name = COUNTRY_NAMES[country]
            
            observations = result["observations"]

            year_to_pop = {
                int(obs["date"]): obs["value"]
                for obs in observations
                if obs["value"] is not None
            }

            for year in years: 
                out.append(
                    {
                        "question": f"What was {c_name}'s population in {year}?",
                        "answer": year_to_pop[year],
                        "unit": "people",
                        "source": "World Bank Open Data (WDI API)",
                        "year": year,
                        "sourceUrl": result["sourceUrl"],
                        "retrievedAt": datetime.utcnow().isoformat(),
                    }
                )

        return out
    
       
    
```
